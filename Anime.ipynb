```python
import pandas as pd
import numpy as np
df=pd.read_csv('https://raw.githubusercontent.com/Mayank-Bhatia/Anime-Recommender/refs/heads/master/data/anime.csv')
df
df.head()
```

```python
df_features=df[['name','genre','type','rating']].fillna('')
df_features
```

```python
x_test=df_features['name'].astype(str)+df_features['genre'].astype(str)+df_features['type'].astype(str)+df_features['rating'].astype(str)
#The error was caused due to rating being a float value. So, we changed the type to string before concatenating.
x_test
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer()
x_tfidf=tfidf.fit_transform(x_test)
print(x_tfidf)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(x_tfidf)
similarity
```

```python
fav_movie=input('Enter your favourite Anime:')
```

```python
all_movies=df['name'].tolist()
import difflib
movie_close_match=difflib.get_close_matches(fav_movie,all_movies)
print(movie_close_match)
```

```python
close_match=movie_close_match[0]
print(close_match)
```

```python
index_of_close_match = df[df['name'] == close_match]['anime_id'].values[0]
print(index_of_close_match)
```

```python
recommendation_score=list(enumerate(similarity[index_of_close_match]))
print(recommendation_score)
```

```python
len(recommendation_score)
```

```python
sorted_similar_movies=sorted(recommendation_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_movies)
```

```python
print("similar movies")
i=1
for movie in sorted_similar_movies:
  index=movie[0]
  title_from_index=df[df.index==index]['name'].values[0]
  if(i<31):
    print(i,'.',title_from_index)
    i+=1
```
